Урок 2. Расчет и визуализация маркетинговых показателей, часть 2


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(24, 12))

# График ежедневных трат на рекламу
plt.subplot(2, 2, 1)
plt.plot(df_cohort_spend['install_date'], df_cohort_spend['spend_daily'], marker='o')
plt.title('Ежедневные траты на рекламу')
plt.xlabel('Дата установки')
plt.ylabel('Траты, $')

# График прибыли
plt.subplot(2, 2, 2)
plt.plot(df_cohort_spend['install_date'], df_cohort_spend['profit'], marker='o', color='green')
plt.title('Прибыль')
plt.xlabel('Дата установки')
plt.ylabel('Прибыль, $')

# График стоимости за установку (CPI)
plt.subplot(2, 2, 3)
plt.plot(df_cohort_spend['install_date'], df_cohort_spend['cpi'], marker='o', color='orange')
plt.title('CPI (стоимость за установку)')
plt.xlabel('Дата установки')
plt.ylabel('CPI, $')

# График ROI
plt.subplot(2, 2, 4)
plt.plot(df_cohort_spend['install_date'], df_cohort_spend['roi'], marker='o', color='red')
plt.title('ROI (Return on Investment)')
plt.xlabel('Дата установки')
plt.ylabel('ROI, %')

plt.tight_layout()
plt.show()
```
